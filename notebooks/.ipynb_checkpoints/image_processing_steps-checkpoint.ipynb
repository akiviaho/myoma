### Tiling

```python
def crop(infile,height,width):
    im = Image.open(infile)
    imgwidth, imgheight = im.size
    for i in range(imgheight//height):
        for j in range(imgwidth//width):
            box = (j*width, i*height, (j+1)*width, (i+1)*height)
            yield im.crop(box)      
   
                      
if __name__ == '__main__':

    filelist = glob('../Slides/Box4/My_6075_T1.mrxs')
    filelist.sort()
    
    for fl in filelist:
        height    = 1000
        width     = 357
        start_num = 1
        
        name     = os.path.basename(fl).replace(".mrxs","")
        src_dir  = os.path.join(os.path.dirname(fl))
        dst_dir  = os.path.join(src_dir + '/' + name)
        
        for k,piece in enumerate(crop(fl,height,width), start_num):
            img=Image.new('RGB', (height,width), 255)
            img.paste(piece)
            path=os.path.join(dst_dir, "%s-%s.png" % (name, k))
            img.save(path)               
```
