```python
import os
import sys
from logzero import logger
from matplotlib import pyplot as plt
from itertools import cycle

import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf

from keras.models import load_model

from sklearn.metrics import auc
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Imported to a random notebook just to see how how the training weights settle

# Sample labeling scheme for string to enum
# Import in all scripts necessary
class toIntEnum(IntEnum):
    wt = 0
    MED12 = 1
    HMGA2 = 2
    YEATS4 = 3

def calculate_class_weights(df):
    weights = []
    cols = np.unique(df['Label'])
    for s in cols:
        weights.append(len(df)/np.where(df['Label'] == s)[0].shape[0])
    weights_dict = dict()
    for i in range(len(weights)):
        weights_dict[cols[i]] = np.round(weights[i]/sum(weights),3)
    return weights_dict

# Calculate class weights
sample_sheet = pd.read_csv('/lustre/scratch/kiviaho/myoma/myoma-new/sample_information_KF_1_2_3_4_in_5.tsv')

# Convert sample sheet to binary label
sample_sheet.replace([2,3],1,inplace=True) # Modify list values to replace anything else    

training_weights = calculate_class_weights(sample_sheet)
N_TARGETS = len(np.unique(sample_sheet['Label'])) # Binary 
```

```python
model_name = 'two-labels-InceptionV3_40_epochs_5_percent_21118730_at_2022-03-31_16:08:12_598px.h5.prediction_results.csv'
results = pd.read_csv('/lustre/scratch/kiviaho/myoma/myoma-new/'+model_name)
results = results[:631320]
gt_tile = results['Label']
results = results.rename(columns={'0': 0, '1': 1})


pred = np.array(results[[0,1]],dtype=np.float32)
```

```python
# get predictions into single column
# for confusion matrix
predictions = np.argmax(pred, axis=1)

# probability for ROC
probs = results.loc[:,[0,1]]

target_names = ['wt','mut']
logger.debug("Confusion Matrix")
cm = confusion_matrix(gt_tile, predictions)
logger.info(cm)

logger.debug("Classification Report")
cr = classification_report(gt_tile, predictions,
                            target_names=target_names)
logger.info(cr)

## Add ROC and AUC
auc_cm = roc_auc_score(gt_tile, probs[1])
logger.info(auc_cm)

fpr, tpr, thresholds = roc_curve(gt_tile, probs[1])
# model_auc = auc(fpr, tpr)

filename = str(model_name).replace('results/', '')
filename = filename.replace('_598px.h5', '')
filename = filename.replace('.prediction_results.csv', '')

fig, (ax1, ax2) = plt.subplots(2,1)
akws = {"ha": 'left',"va": 'center'}
sns.heatmap(cm, annot=True, ax = ax1, fmt='g', annot_kws=akws)
# labels, title and ticks
ax1.set_xlabel('Predicted labels')
ax1.set_ylabel('True labels')
ax1.set_title('Confusion Matrix')
ax1.xaxis.set_ticklabels(target_names)
ax1.yaxis.set_ticklabels(target_names)

ax2.plot([0, 1], [0, 1], 'k--')
ax2.plot(fpr, tpr, label='Prediction (area = {:.3f})'.format(auc_cm))
ax2.set_xlabel('False positive rate')
ax2.set_ylabel('True positive rate')
ax2.set_title('ROC curve')
ax2.legend(loc='best')
fig.suptitle(filename, fontsize=12)
fig.set_size_inches(8, 12)
logger.info("Saving classification report plots")
fig.savefig(filename + '.pdf', dpi=300)

```

```python
# ADD ONE INSTANCE OF CLASS 4
# OTHERWISE ROC/AUC CALCULATIONS FAIL
# PREPROCESS THE ARRATS TO MATCH 

n_classes = 5
target_names = ['wild type', 'SNV44','DELINS','SNV36','SNV43']
lw = 1

#gt_tile_mod = np.append(gt_tile,4)
#results_mod = results.append(pd.DataFrame({'foo':['foo','foo',4,1,0,0,0,0]},index=['Sample','Path','Label',0,1,2,3,4,]).transpose())

results_mod = results
gt_tile_mod = np.squeeze(np.array(gt_tile,dtype=np.int16))

gt_tile_mod = np.array(tf.one_hot(gt_tile_mod,n_classes),dtype=np.int16)
results_mod = np.array(results_mod.loc[:,[0,1,2,3,4]],dtype=np.float32)
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(gt_tile_mod[:, i], results_mod[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(gt_tile_mod.ravel(), results_mod.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
plt.figure()

for i in range(n_classes):
    plt.plot(
        fpr[i],
        tpr[i],
        lw=lw,
        label= target_names[i] + " (area = %0.2f)" % roc_auc[i],
    )
plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver operating characteristic example")
plt.legend(loc="lower right")
plt.show()
```

```python
# First aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

# Then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

# Finally average it and compute AUC
mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot all ROC curves
plt.figure()
plt.plot(
    fpr["micro"],
    tpr["micro"],
    label="Average ROC curve (area = {0:0.2f})".format(roc_auc["micro"]),
    color="deeppink",
    linestyle=":",
    linewidth=4,
)


colors = cycle(["aqua", "darkorange", "cornflowerblue",'violet'])
for i, color in zip(range(n_classes-1), colors):
    plt.plot(
        fpr[i],
        tpr[i],
        color=color,
        lw=lw,
        label= target_names[i] + " (area = %0.2f)" % roc_auc[i],
    )

plt.plot([0, 1], [0, 1], "k--", lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Tile Level Receiver Operating Charasteristic")
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig(filename + '_ROC_tile_level.png', dpi=300)
```
