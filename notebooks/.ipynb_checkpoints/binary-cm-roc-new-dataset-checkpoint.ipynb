```python
import os
import sys
from logzero import logger
from matplotlib import pyplot as plt
from itertools import cycle
from scipy import stats

import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
from enum import IntEnum

from matplotlib.colors import LogNorm, Normalize
from matplotlib.ticker import MaxNLocator

from keras.models import load_model

from sklearn.metrics import auc
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

class toIntEnum(IntEnum):
    wt = 0
    MED12 = 1
    HMGA2 = 2
    YEATS4 = 3
```

```python
# Imported to a random notebook just to see how how the training weights settle

def calculate_class_weights(df):
    weights = []
    cols = np.unique(df['Label'])
    for s in cols:
        weights.append(len(df)/np.where(df['Label'] == s)[0].shape[0])
    weights_dict = dict()
    for i in range(len(weights)):
        weights_dict[cols[i]] = np.round(weights[i]/sum(weights),3)
    return weights_dict

def plot_random_tiles(df,n_tiles):
    plt.figure(figsize=(n_tiles, n_tiles))

    i= 0
    for n in np.random.randint(0,len(df),n_tiles):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(plt.imread(df['Tile'][n])) 
        plt.title(
            df['Sample'][n]
                 )
        plt.axis("off")
        i+=1
    return

# Calculate class weights
sample_sheet = pd.read_csv('/lustre/scratch/kiviaho/myoma/myoma-new/sample_information_KF_1_2_3_4_in_5.tsv',sep='\t')

# Convert sample sheet to binary label
sample_sheet.replace([2,3],1,inplace=True) # Modify list values to replace anything else    
labels = [ toIntEnum[s].value for s in sample_sheet['Type'] ]
sample_sheet['Label'] = labels

training_weights = calculate_class_weights(sample_sheet)
N_TARGETS = len(np.unique(sample_sheet['Label'])) # Binary 
```

```python
tfrecord_contents = pd.read_csv('/lustre/scratch/kiviaho/myoma/myoma-new/KF_5_in_5_tfr_contents_in_order.tsv')

# Convert labels to 0-1 binary (wt/mut)
tfrecord_contents.replace([2,3],1,inplace=True) 
```

```python
res_sheet = 'binary_label_KF-1-4_1_epochs_100_percent_21875129_at_2022-06-12_18:46:52_598px.h5_new_dataset_binary_prediction_results.csv'
results = pd.read_csv('/lustre/scratch/kiviaho/myoma/myoma-new/'+res_sheet)
results.dropna(inplace=True)

# Convert labels to 0-1 binary (wt/mut)
results.replace([2,3],1,inplace=True) 
tfrecord_contents = tfrecord_contents[:len(results)]


gt_tile = tfrecord_contents['Label']
results = results.rename(columns={'0': 0, '1': 1})

pred = np.array(results[[0,1]],dtype=np.float32)
```

```python
# get predictions into single column
# for confusion matrix
probs = results.loc[:,[0,1]]

predictions = probs.idxmax(axis=1)

cm = confusion_matrix(gt_tile, predictions)
logger.info(cm)

target_names = ['Wild Type  \n ({:.2e})'.format(np.sum(cm[0,:])),'Mutated  \n ({:.2e})'.format(np.sum(cm[1,:]))]
predict_names = ['Wild Type','Mutated']

logger.debug("Classification Report")
cr = classification_report(gt_tile, predictions,
                            target_names=target_names,digits=3)
logger.info(cr)

## Add ROC and AUC
auc = roc_auc_score(gt_tile, probs[1])
logger.info(auc)

# Switch to percentage. Remember to switch sns.heatmap(fmt='.2e' to fmt='.2%')
cm = cm/cm.sum(axis=1, keepdims=True)


ax1 = plt.subplot(111)
akws = {"ha": 'left',"va": 'center'}

sns.set(font_scale=1.2)
sns.heatmap(cm, annot=cm, ax = ax1, fmt='.2f', annot_kws=akws,linewidths=2,
            cbar =True, linecolor='white', vmin=0.,vmax=1.)

# labels, title and ticks
ax1.set_xlabel('Predicted labels')
ax1.set_ylabel('True labels')
ax1.set_title("Tile Level")
ax1.xaxis.set_ticklabels(predict_names)
ax1.yaxis.set_ticklabels(target_names)
ax1.figure.set_size_inches((12,8))
ax1.figure.set_dpi(200)

#plt.savefig(filename + '.png', dpi=300)
```

```python

## Add ROC and AUC
auc_cm = roc_auc_score(gt_tile, probs[1])
logger.info(auc_cm)

fpr, tpr, thresholds = roc_curve(gt_tile, probs[1])
# model_auc = auc(fpr, tpr)

plt.style.use('fast')
ax2 = plt.axes()
ax2.plot([0, 1], [0, 1], 'k--')
ax2.plot(fpr, tpr,'r-', label='Prediction (area = {:.2f})'.format(auc_cm),linewidth=3)
ax2.set_xlabel('False positive rate',size=15)
ax2.set_ylabel('True positive rate',size=15)
ax2.set_title('Tile Level ROC Curve')
ax2.legend(loc='best',prop={'size': 15})
ax2.figure.set_size_inches((10,8))
ax2.figure.set_dpi(200)
```
