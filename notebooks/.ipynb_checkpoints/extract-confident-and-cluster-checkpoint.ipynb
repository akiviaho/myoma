```python
from matplotlib import pyplot as plt
import seaborn as sns
from itertools import cycle

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from collections import Counter
from scipy.stats import fisher_exact
from numpy.random import default_rng

import warnings
warnings.filterwarnings('ignore')

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE

import time

from enum import IntEnum
class toIntEnum(IntEnum):
    MED12 = 0
    HMGA2 = 1
    UNK = 2
    HMGA1 = 3
    OM = 4
    YEATS4 = 5

```

```python
def plot_random_tiles(df,n_tiles):
    plt.figure(figsize=(n_tiles*2, n_tiles*2))
    i= 0
    rng = default_rng()
    for n in rng.choice(len(df), size=n_tiles, replace=False):
        ax = plt.subplot(5, 5, i + 1)
        plt.imshow(plt.imread(df['Path'][n])) 
        plt.title(
            df['Sample'][n] + ': ' + toIntEnum(df['Label'][n]).name,
                 fontsize=25)
        plt.axis("off")
        i+=1
    return
```

```python
which_fold = 'fold_4'
metadata = pd.read_csv('/lustre/scratch/kiviaho/myoma/myoma-new/conv2d_93_layers/'+which_fold+'_confidently_predicted.tsv',sep='\t')
data = np.load('/lustre/scratch/kiviaho/myoma/myoma-new/conv2d_93_layers/'+which_fold+'_50_PCs_data.npy')
save_path = '/lustre/scratch/kiviaho/myoma/myoma-new/subclass-features/'

data = np.transpose(data)

n_clusters = 10
OR_thresh = 5
p_val_thresh = 0.001
```

```python
data.shape
```

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(data)
metadata['Cluster'] = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
valid_clusters = []
cluster_types = []

for i in range(n_clusters):
    labels_cluster = [toIntEnum(f).name for f in metadata[metadata['Cluster'] == i]['Label']]
    cntr = Counter(labels_cluster)
    vals = list(cntr.values())
    
    cluster_top_key = list(cntr.keys())[np.argmax(vals)]
    cluster_top_value = np.max(list(cntr.values()))
    cluster_n = len(labels_cluster)
    
    labels_all = np.array([toIntEnum(f).name for f in metadata['Label']])
    n = len(labels_all)
    cnt_all = Counter(labels_all)
    top_sum = cnt_all[cluster_top_key]
    
    
    a = n - top_sum
    b = cluster_n - cluster_top_value
    c = top_sum 
    d = cluster_top_value
    
    cont_table = np.array([[a,b],[c,d]])
    fshr_res = fisher_exact(cont_table)
    if fshr_res[0] > OR_thresh and fshr_res[1] < p_val_thresh:
        valid_clusters.append(i)
        cluster_types.append(toIntEnum[cluster_top_key].value)
    
#     print('Cluster: ' + str(i))
#     print(cluster_top_key)
#     print(cont_table)
#     print(fshr_res)
#     print(' ')
```

```python
print(valid_clusters)
print(cluster_types)
```

```python
valid_idxs = []
for c,l in zip(valid_clusters,cluster_types):
    idxs = metadata[(metadata['Cluster'] == c) & (metadata['Label']  ==  l)].index
    for i in idxs:
        valid_idxs.append(i)
```

```python
valid_idxs = np.asarray(valid_idxs)
metadata_to_save = metadata.iloc[valid_idxs].reset_index(drop=True)
```
